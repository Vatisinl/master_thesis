```r
colnames(summary(df_top, type = "A3SS"))
colnames(summary(df_top, type = "A5SS"))
colnames(summary(df_top, type = "MXE"))
colnames(summary(df_top, type = "SE"))
colnames(summary(df_top, type = "RI"))
#summary(df_top, type = "SE")

RI <- summary(df_top, type = "RI")
SE <- summary(df_top, type = "SE")
A3SS <- summary(df_top, type = "A3SS")
A5SS <- summary(df_top, type = "A5SS")
MXE <- summary(df_top, type = "MXE")

RI$type <- 'RI'
SE$type <- 'SE'
A3SS$type <- 'A3SS'
A5SS$type <- 'A5SS'
MXE$type <- 'MXE'
```

```r
#start for A3SS
start <- strtoi(trimws(summary(df_top, type = "A3SS")$exon_long, whitespace = "-.*"))
A3SS$start <- start
#end for A3SS
end <- strtoi(trimws(summary(df_top, type = "A3SS")$exon_flanking, whitespace = ".*-"))
A3SS$end <- end

#start for A5SS
start <- strtoi(trimws(summary(df_top, type = "A5SS")$exon_flanking, whitespace = "-.*"))
A5SS$start <- start
#end for A5SS
end <- strtoi(trimws(summary(df_top, type = "A5SS")$exon_long, whitespace = ".*-"))
A5SS$end <- end

#start for SE
start <- strtoi(trimws(summary(df_top, type = "SE")$exon_upstream, whitespace = "-.*"))
SE$start <- start
#end for SE
end <- strtoi(trimws(summary(df_top, type = "SE")$exon_downstream, whitespace = ".*-"))
SE$end <- end

#start for MXE
start <- strtoi(trimws(summary(df_top, type = "MXE")$exon_upstream, whitespace = "-.*"))
MXE$start <- start
#end for MXE
end <- strtoi(trimws(summary(df_top, type = "MXE")$exon_downstream, whitespace = ".*-"))
MXE$end <- end

#start for RI
start <- strtoi(trimws(summary(df_top, type = "RI")$exon_upstream, whitespace = "-.*"))
RI$start <- start
#end for 
end <- strtoi(trimws(summary(df_top, type = "RI")$exon_downstream, whitespace = ".*-"))
RI$end <- end
```
